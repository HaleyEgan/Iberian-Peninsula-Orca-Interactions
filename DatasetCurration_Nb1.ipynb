### Overview of Dataset and Variables
** All empty columns were removed

** Columns not needed in our dataset are removed.

- 'dataset_id': The dataset_id is '178820dc-ead7-497b-b2a2-2c8ea4d5a084', and is the same for each entry (same dataset). - Column Removed
- 'id': unique id for each entry
- 'aphiaid': value '137102' occurs for every entry (107 times) - Column removed
- 'areas': unclear what variable signifies. Mostly the same throughout - Column removed
- 'associatedMedia': provides a unique link for each entry, which leaved to anavnet, with what appears to be the mapped location of each siting/interaction
- 'basisOfRecord': All entries record 'HumanObservation' as the method for observation - Column Removed
- 'bathymetry': Water depth at a location (often in meters, negative down)
- 'catalogNumber': Each entry has a unique catalogNumber - Column Removed (same as occurrenceID)
- 'class': All entries report class as 'Mammalia'. - Column Removed
- 'classid': All entries have 'classid' '1837' - Column Removed
- 'datasetID': All values are 'https://marineinfo.org/id/dataset/8743' - Column Removed
- 'datasetName': All values are 'Events with orcas - AnavNet' - Column Removed
- 'date_end': Unclear what this data represents, but doesn't appear to be relevant to our needs - Column Removed
- 'date_mid': Unclear what this data represents, but doesn't appear to be relevant to our needs - Column Removed
- 'date_start': Unclear what this data represents, but doesn't appear to be relevant to our needs - Column Removed
- 'date_year': Year of recorded encounter. But this column is identical to the column 'year', so this one is dropped and 'year' kept - Column Removed
- 'day': Day of the month of encounter. 
- 'decimalLatitude': Latitude coordinates of event location (float)
- 'decimalLongitude': Longitude coordinates of event location (float)
- 'eventDate': date of event, yyyy-mm
- 'family': All values are of family 'Delphinidae' - Column Removed
- 'familyid': All values are of familyid '136980' - Column Removed
- 'footprintWKT': Geospatial data in WKT (Well-Known Text) format. A single point location expressed as a MultiPoint, with latitude and longitude.
- 'genus': All values for genus are 'Orcinus' - Column Removed
- 'genusid': All values are '137021' - Column Removed
- 'infraorder': All values are 'Cetacea' for order. - Column Removed

- 'occurrenceStatus': = 'present' - Column Removed
